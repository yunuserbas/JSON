```python
import json
import requests

link = "https://raw.githubusercontent.com/murpi/wilddata/master/quests/monthlySalesbyCategoryMultiple.json"

# This function gets the file content from the URL
r = requests.get(link)


# You can access to the text content with this function
r.text
```

```python
data = json.loads( r.text)
data
```

```python
data['contents'][1]['monthlySales']
```

```python
data['contents'][0]['category']
```

```python
import pandas as pd

# You can directly read JSON from your request object
print(r.json())

# And you can import JSON object to pandas
pd.json_normalize(r.json())

# You can scroll with the record_path argument
# and keep some information with meta argument
```

```python
pd.json_normalize(data['contents'], 'monthlySales',['category', 'region'])
```
